```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import requests

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time

def most_viewed_videos_wiki(url):

    opts = Options()
    driver = webdriver.Chrome(r"C:\Users\Admin\.wdm\drivers\chromedriver\win32\87.0.4280.88\chromedriver.exe",options=opts)

    url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"

    opts.add_argument("user-agent=Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36")

    driver.get(url)

    driver.find_element_by_xpath("//b").click()

    time.sleep(3)

    records = []
    for row in driver.find_elements_by_xpath("(//table[@class='wikitable sortable jquery-tablesorter'])[1]//tr")[1:-1]:
        temp = ''
        for element in row.find_elements_by_xpath(".//td"):        
            temp += element.text + '|'
        records.append(temp.split('|')[:-2])

    df = pd.DataFrame(records,columns=['rank','video_name','artist','views(M)','date_uploaded'])
    
    arranged_cols = ['rank','video_name','artist','date_uploaded','views(M)']
    
    df.video_name = df.video_name.apply(lambda x:x[:-4].strip('"'))
    df = df[arranged_cols]

    return df


most_viewed_videos_wiki("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import requests

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.action_chains import ActionChains

import time


def india_fixtures(url):

    opts = Options()
    driver = webdriver.Chrome(r"C:\Users\Admin\.wdm\drivers\chromedriver\win32\87.0.4280.88\chromedriver.exe",options=opts)

    opts.add_argument("user-agent=Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36")


    driver.get(url)
    
    tab_hover = driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover']")

    ActionChains(driver).move_to_element(tab_hover).perform()

    driver.find_element_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down']").click()
    
    time.sleep(2)

    title_l,series_l,place_l,date_l,time_l = [],[],[],[],[]

    for date in driver.find_elements_by_xpath("//div[@class='fixture__datetime desktop-only']"):
        date_l.append(' '.join(date.text.replace('\n',' ').split(' ',3)[:-1]))
        time_l.append(date.text.replace('\n',' ').split(' ',3)[-1])
    for title in driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']"):
        title_l.append(title.text)
    for series in driver.find_elements_by_xpath("//div[@class='fixture__teams']"):
        series_l.append(series.text.replace('\n',' '))
    for place in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span"):
        place_l.append(place.text)


    df = pd.DataFrame(list(zip(title_l,series_l,place_l,date_l,time_l)),columns=['title','series','place','date','time'])

    return df

india_fixtures('https://www.bcci.tv/')
```

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import requests

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException

def trending_git(url):
    opts = Options()
    driver = webdriver.Chrome(r"C:\Users\Admin\.wdm\drivers\chromedriver\win32\87.0.4280.88\chromedriver.exe",options=opts)

    opts.add_argument("user-agent=Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36")

    title_l,des_l,con_count_l,lang_l = [],[],[],[]
    
    driver.get(url)
    
    time.sleep(2)
    
    driver.find_elements_by_xpath("//summary[@class='HeaderMenu-summary HeaderMenu-link px-0 py-3 border-0 no-wrap d-block d-lg-inline-block']")[1].click()
        
    driver.find_element_by_xpath("//a[contains(@data-ga-click,'Trending')]").click()
    
    
    for block in driver.find_elements_by_xpath("//article[@class='Box-row']"):
        title_l.append(block.find_element_by_xpath(".//h1[@class='h3 lh-condensed']").text)
        des_l.append(block.find_element_by_xpath(".//p[@class='col-9 text-gray my-1 pr-4']").text)
        con_count_l.append(block.find_elements_by_xpath(".//a[@class='muted-link d-inline-block mr-3']")[1].text)
        try:
            lang_l.append(block.find_element_by_xpath(".//span[@itemprop='programmingLanguage']").text)
        except NoSuchElementException:
            lang_l.append('NA')    
        
    df = pd.DataFrame(list(zip(title_l,des_l,con_count_l,lang_l)),columns=['title','description','countributors_count','language'])
    
    return df

trending_git("https://github.com/")
```

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import requests

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

import time

def data_scientist_scrapes(url):

    opts = Options()
    driver = webdriver.Chrome(r"C:\Users\Admin\.wdm\drivers\chromedriver\win32\87.0.4280.88\chromedriver.exe",options=opts)

    opts.add_argument("user-agent=Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36")

    driver.get(url)

    driver.find_element_by_id('qsb-keyword-sugg').send_keys('Data Science')

    driver.find_element_by_class_name('btn').click()
    
    time.sleep(2)
    
    # scraping the data:

    job_titles, job_company = [],[]

    job_location,job_skills  = [],[]

    for title in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
        job_titles.append(title.text)

    for company_name in driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']"):
        job_company.append(company_name.text)
        
    for skills in driver.find_elements_by_xpath("//ul[@class='tags has-description']"):
        temp = skills.text
        job_skills.append(temp.replace('\n','|'))     
        
    for i,value in enumerate(driver.find_elements_by_xpath("//span[@class='ellipsis fleft fs12 lh16']")):
        if i%3 == 2:
            job_location.append(value.text)
    
    df = pd.DataFrame({'job_title':job_titles,'job_company':job_company,'job_locations':job_location,'job_skills':job_skills})

    return df

data_scientist_scrapes('https://www.naukri.com/')
```
